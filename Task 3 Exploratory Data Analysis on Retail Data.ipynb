### Univariate Analysis

```python
for i in num_cols:
    plt.figure(figsize=(6,3))
    plt.style.use("tableau-colorblind10")
    sns.distplot(df[i], bins = 50, color = 'green')
    plt.xlabel(i)
    plt.title("Frequency Distribution of" +i)
    plt.xticks(rotation = 90)
    plt.show()
    
```

#### Distplot represent the univariate distribution of data:
###### * Data distribution of a variable against density distribution

```python
### let's use countplot for categorical feature

for i in cat_cols:
    plt.figure(figsize=(8,5))
    sns.countplot(df[i])
    plt.xlabel(i)
    plt.title("Count Distribution of" +i)
    plt.xticks(rotation = 90)
    plt.show()
```

### Using Boxplot for detecting outliers

```python
# cheking outliers in Sales feature

sns.boxplot(y = 'Sales', data = df)
```

```python
# Removing outilier in Sales Features


q1 = df['Sales'].quantile(0.25)
q3 = df['Sales'].quantile(0.75)
iqr = q3-q1

#low_limit = q1-1.5*iqr
upper_limit = q3+1.5*iqr
upper_limit
```

```python
df['Sales']=np.where(df['Sales'] > upper_limit,upper_limit,df['Sales'])
```

```python
sns.boxplot(y = 'Quantity' , data = df)
```

```python
# Removing outilier in Quantity Features

q1 = df['Quantity'].quantile(0.25)
q3 = df['Quantity'].quantile(0.75)
iqr = q3-q1

#low_limit = q1-1.5*iqr
upper_limit = q3+1.5*iqr
upper_limit
```

```python
df['Quantity']=np.where(df['Quantity'] > upper_limit,upper_limit,df['Quantity'])
```

```python
# checking outliers in Discount features

sns.boxplot(y = 'Discount' , data = df)
```

```python
# Removing outliers in Discount features

q1 = df['Discount'].quantile(0.25)
q3 = df['Discount'].quantile(0.75)
iqr = q3-q1
#low_limit = q1-1.5*iqr
upper_limit = q3+1.5*iqr
upper_limit
```

```python
df['Discount']=np.where(df['Discount'] > upper_limit,upper_limit,df['Discount'])
```

```python
# checking outliers in Profit features

sns.boxplot(y = 'Profit' , data = df)
```

```python
# Removing outliers in Profit features

q1 = df['Profit'].quantile(0.25)
q3 = df['Profit'].quantile(0.75)
iqr = q3-q1
lower_limit = q1-1.5*iqr
upper_limit = q3+1.5*iqr
upper_limit
```

```python
df['Profit']=np.where(df['Profit'] > upper_limit,upper_limit,df['Profit'])
df['Profit']=np.where(df['Profit'] < lower_limit,lower_limit,df['Profit'])
```
